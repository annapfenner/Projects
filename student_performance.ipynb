```python
def removal_box_plot(df, column, threshold):
    sns.boxplot(df[column])
    plt.title(f'Original Box Plot of {column}')
    plt.show()
 
    removed_outliers = df[df[column] >= threshold]
 
    sns.boxplot(removed_outliers[column])
    plt.title(f'Box Plot without Outliers of {column}')
    plt.show()
    return removed_outliers
 
df = removal_box_plot(data, 'math_score', 20)
```

## 5. Pre-processing

```python
print(data.race_ethnicity.value_counts())
print(data.parental_level_of_education.value_counts())
print(data.test_preparation_course.value_counts())
```

```python
df = pd.get_dummies(data, columns = ['gender', 'race_ethnicity','parental_level_of_education', 'test_preparation_course', 'lunch'])
df.head()
```

```python
y = df.math_score
X = df.drop('math_score', axis =1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=47)
```

## 6. Modeling

### 6a. Plots to look at distribution of residuals and see there is no pattern to it aka no linearity to data. 

```python
import statsmodels.api as sm

# Create a linear regression model using statsmodels
X = sm.add_constant(X)  
model = sm.OLS(y, X).fit()

# Plot Q-Q plot of residuals
sm.qqplot(model.resid, line='s')
plt.title('Q-Q Plot of Residuals')
plt.show()

# Plot residuals vs. fitted values
plt.scatter(model.fittedvalues, model.resid)
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Fitted Values')
plt.show()
```

### 6b. Linear Regression

```python
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)
```

```python
y_pred= model_lr.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
r_2_value = r2_score(y_test, y_pred)
print("R^2 value:", r_2_value)
```

### 6c. XGBoost

```python
import xgboost as xgb
xgb_model = xgb.XGBRegressor()

# Fit the model on the training data
xgb_model.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = xgb_model.predict(X_test)

# Evaluate the model using mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

r_2_value = r2_score(y_test, y_pred)
print("R^2 value on test set:", r_2_value)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBRegressor
from scipy.stats import uniform, randint

# Define the parameter grid
param_grid = {
    'learning_rate': uniform(0.01, 0.3), 
    'n_estimators': randint(100, 1000),   
    'max_depth': randint(3, 10),          
    'subsample': uniform(0.5, 0.5),       
    'colsample_bytree': uniform(0.5, 0.5),
    'gamma': uniform(0, 0.5),             
    'reg_alpha': uniform(0, 1),          
    'reg_lambda': uniform(0, 1),          
}

# Create an instance of the XGBRegressor model
xgb_model = XGBRegressor()

# Create RandomizedSearchCV object
random_search = RandomizedSearchCV(estimator=xgb_model, param_distributions=param_grid,
                                   n_iter=100, scoring='neg_mean_squared_error', cv=3,
                                   verbose=1, random_state=42)

# Fit RandomizedSearchCV to the data
random_search.fit(X_train, y_train)
```

```python
# Get the best parameters
print("Best parameters found: ", random_search.best_params_)

# Get the best score
print("Best RMSE on validation data: ", (-random_search.best_score_) ** 0.5)

# Evaluate the best model on test data
best_model = random_search.best_estimator_
y_pred = best_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error on test data:", mse)

r_2_value = r2_score(y_test, y_pred)
print("R^2 value on test set:", r_2_value)
```
